```python
category_subscriber_mean = df.groupby('category')['subscribers'].mean()
category_subscriber_mean
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
category_subscriber_mean .sort_values(ascending = False)
```

```python
category_subscriber_mean.plot(kind = "bar",color = "red")
plt.title("Mean subscriber by category")
plt.xlabel("category")
plt.ylabel("Mean subscriber")
plt.xticks(rotation = 90)
plt.show()
```

```python
channel_type_counts = df["channel_type"].value_counts()
channel_type_counts
```

```python
plt.pie(channel_type_counts.head(), labels=channel_type_counts.head().index, autopct = "%1.1f%%", colors = sns.color_palette("coolwarm"), wedgeprops={"edgecolor":"black"})
plt.title("Channel Type Distribution")
plt.show()
```

```python
plt.scatter(df["subscribers"], df["video views"])
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.title("scatter plot of subscribers vs video views")
plt.show()
```

```python
country_subscribers_total = df.groupby('Country')["subscribers"].sum().sort_values(ascending = False)
```

```python
country_subscribers_total.head(10).plot(kind = "bar", colormap = "viridis")
plt.xlabel("country")
plt.ylabel("Total Subscribers")
plt.title("Total Subscriber by Country")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(x = "channel_type", y = "highest_monthly_earnings",data = df, palette = "rainbow")
plt.title("distribution of highest monthly earning by channel type")
plt.xticks(rotation = 90)
plt.show()
```

```python
df.groupby(["category","Country"])["subscribers"].sum().unstack().plot(kind = "bar",stacked = True)
plt.title("subscribers by category and country")
plt.xlabel("category")
plt.ylabel("subscribers")
plt.legend(title = "Country",bbox_to_anchor = (1.05, 1, 0., 0.), loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"], df["subscribers"],alpha = 0.5)
plt.title("scatter plot of Subscribers vs gross tertiary education enrollment")
plt.xlabel("Gross tertiary education enrollment")
plt.ylabel("subscribers")
plt.show()
```

```python
ax = df.groupby("channel_type")["Unemployment rate"].mean().plot(kind = 'barh')
for bar in ax.containers:
    ax.bar_label(bar, fmt='%.2f')
plt.title("Average Unemployment Rate by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Average Unemployment Rate")
plt.show()
```

```python
ax = df.groupby(["created_year","channel_type"])["subscribers"].mean().unstack().tail().plot(kind = "bar")
plt.title("Average subscriber by created year and channel type")
plt.xlabel("created year")
plt.ylabel("Average_subscriber")
plt.xticks(rotation = 45)
plt.legend(title = "chennel type",bbox_to_anchor = (1.05, 1, 0., 0.),loc = "upper left")
plt.show()
```

```python
df.groupby(["created_year","created_month"])["subscribers"].sum().unstack().head().plot(kind = "area", stacked = True)
plt.title("suscribers growth over time")
plt.xlabel("year")
plt.ylabel("total subscriber")
plt.legend(title = "month",bbox_to_anchor = (1.05, 1, 0., 0.),loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
df.groupby(["created_year","Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind = "line", marker = "o",linestyle = "-")
plt.title("suscribers growth over time")
plt.xlabel("year")
plt.ylabel("total subscriber")
plt.legend(title = "month",bbox_to_anchor = (1.05, 1, 0., 0.),loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
channel_created_subscriber_mean = df.groupby(["channel_type","created_year"])["subscribers_for_last_30_days"].mean().unstack(fill_value = 0)
```

```python
channel_created_subscriber_mean
```

```python
plt.figure(figsize = (20,8))
sns.heatmap(channel_created_subscriber_mean,cmap = "YlGnBu",annot = True)
plt.title("Mean Subscriber for last 30 days by channel type and created year")
plt.xlabel("created year")
plt.ylabel("channe type")
plt.show()
```

```python
df.groupby("created_month")["subscribers_for_last_30_days"].mean().plot(kind= "line",marker = "o",linestyle = "-")
plt.title("Average subscribers for last 30 days by months")
plt.xlabel("month")
plt.ylabel("Average subscriber for last 30 days")
plt.xticks(rotation = 45)
plt.show()
```
